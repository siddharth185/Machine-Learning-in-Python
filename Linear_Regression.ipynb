```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
bikes = pd.read_csv("bikeshare.csv",index_col='datetime', parse_dates=True)
```

```python
bikes.head()
```

```python
bikes.rename(columns={"count":"total"},inplace=True)
```

```python
x = bikes.plot(kind='scatter', x='temp', y='total', alpha=0.2)
```

```python
feature_cols = ['temp']
X = bikes[feature_cols]
y = bikes.total
```

```python
linreg = LinearRegression().fit(X, y)
linreg
```

```python
print linreg.intercept_
print linreg.coef_
```

```python
x = bikes.plot(kind='scatter', x='temp', y='total', alpha=0.2)
plt.plot(X, linreg.predict(X), color='red',
         linewidth=3)
```

```python
linreg.predict(20)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print mean_squared_error(y, linreg.predict(X))
print r2_score(y, linreg.predict(X))
```
